# **mAP(mean Average Precision)**
ì—¬ëŸ¬ í´ë˜ìŠ¤ì— ëŒ€í•œ AP ê°’ì„ í‰ê·  ë‚¸ ê°’ìœ¼ë¡œ, ì „ì²´ ëª¨ë¸ì˜ ê°ì²´ íƒì§€ ì„±ëŠ¥ì„ í‰ê°€í•˜ëŠ” ì¢…í•©ì ì¸ ì§€í‘œì…ë‹ˆë‹¤. ë‹¨ì¼ í´ë˜ìŠ¤ê°€ ì•„ë‹Œ ì „ì²´ ë°ì´í„°ì…‹ì— í¬í•¨ëœ ëª¨ë“  í´ë˜ìŠ¤ì˜ ì„±ëŠ¥ì„ ë°˜ì˜í•˜ë©°, ë‹¤ì–‘í•œ IoU ì„ê³—ê°’(ì˜ˆ: 0.5~0.95)ì— ëŒ€í•´ í‰ê· ì„ ë‚´ê¸°ë„ í•©ë‹ˆë‹¤. ë†’ì€ mAP ê°’ì€ ëª¨ë¸ì´ ë‹¤ì–‘í•œ í´ë˜ìŠ¤ì˜ ê°ì²´ë¥¼ ì •í™•í•˜ê²Œ íƒì§€í•œë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•©ë‹ˆë‹¤.

### mAP@50
IoU(Intersection over Union)ê°€ 0.5(50%) ì´ìƒì´ë©´ ì •ë‹µìœ¼ë¡œ ì¸ì •í•˜ê³  ê³„ì‚°í•œ í‰ê·  ì •ë°€ë„ì…ë‹ˆë‹¤. ì¦‰, ëª¨ë¸ì´ ëŒ€ëµì ì¸ ìœ„ì¹˜ë¼ë„ ë§ì¶”ë©´ ì„±ëŠ¥ì´ ì¢‹ê²Œ ë‚˜ì˜µë‹ˆë‹¤.

### mAP@50-95
IoUë¥¼ 0.5ë¶€í„° 0.95ê¹Œì§€ 0.05 ê°„ê²©ìœ¼ë¡œ ë³€í™”ì‹œí‚¤ë©´ì„œ ì„±ëŠ¥ì„ í‰ê°€í•œ í‰ê·  ì •ë°€ë„ì…ë‹ˆë‹¤. ì¦‰, ê°ì²´ì˜ ìœ„ì¹˜ë¥¼ ë”ìš± ì •í™•í•˜ê²Œ ë§ì¶°ì•¼ ë†’ì€ ì ìˆ˜ë¥¼ ë°›ê¸° ë•Œë¬¸ì— ë” ì—„ê²©í•œ ê¸°ì¤€ì…ë‹ˆë‹¤.

### FPS
ì´ˆë‹¹ ëª‡ ê°œì˜ ì´ë¯¸ì§€ë¥¼ ì²˜ë¦¬í•  ìˆ˜ ìˆëŠ”ì§€ë¥¼ ì˜ë¯¸í•˜ë©°, ê°’ì´ ë†’ì„ìˆ˜ë¡ ëª¨ë¸ì´ ë¹ ë¥´ê²Œ ë™ì‘í•œë‹¤ëŠ” ëœ»ì…ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´ 30 FPSë¼ë©´ 1ì´ˆì— 30ì¥ì˜ ì´ë¯¸ì§€ë¥¼ ì²˜ë¦¬í•  ìˆ˜ ìˆë‹¤ëŠ” ì˜ë¯¸ì…ë‹ˆë‹¤.

### Latency
í•œ ì¥ì˜ ì´ë¯¸ì§€ë¥¼ ì…ë ¥í–ˆì„ ë•Œ ê²°ê³¼ê°€ ë‚˜ì˜¤ê¸°ê¹Œì§€ ê±¸ë¦¬ëŠ” ì‹œê°„ìœ¼ë¡œ, ë³´í†µ ë°€ë¦¬ì´ˆ(ms) ë‹¨ìœ„ë¡œ ì¸¡ì •í•©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´ Latencyê°€ 50msë¼ë©´, í•œ ì¥ì˜ ì´ë¯¸ì§€ë¥¼ íƒì§€í•˜ëŠ” ë° 0.05ì´ˆê°€ ê±¸ë¦°ë‹¤ëŠ” ëœ»ì…ë‹ˆë‹¤.

```python
model = YOLO("runs/detect/custom_s5/weights/best.pt")

# ê²€ì¦ ìˆ˜í–‰
results = model.val(
    data="custom_voc.yaml",  # ë°ì´í„°ì…‹ ì„¤ì •
    imgsz=640,               # ì´ë¯¸ì§€ í¬ê¸°
    iou=0.5,                 # IoU ì„ê³„ê°’
    batch=32,                # ë°°ì¹˜ í¬ê¸°
    device=0,                # GPU ì‚¬ìš©
    workers=2,               # ë°ì´í„° ë¡œë“œ ì‹œ ë³‘ë ¬ ì²˜ë¦¬í•  ì›Œì»¤ ìˆ˜
    half=True,               # FP16 ì—°ì‚° í™œì„±í™” (ì†ë„ í–¥ìƒ) 'ì†Œìˆ«ì ì„ ì§§ê²Œ ê°€ì ¸ê°€ë¼'
    split="test"             # í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ì„ ì‚¬ìš©
)

print(results)
```
